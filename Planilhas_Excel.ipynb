**Trabalhando com Planilhas do Excel**

```python
#importanto a biblioteca
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
df4.head()
```

```python
df5.head()
```

```python
#Juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Exibindo as 5 primeiras linhas
df.head()
```

```python
#Exibindo as 5 últimas linhas
df.tail()
```

```python
#Amostra de dados
df.sample(5)
```

```python
#Verificando o tipo de dados de cada coluna
df.dtypes
```

```python
#Alterando o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

**Tratando Valores faltantes**

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplane=True)
```

```python
df.isnull().sum()
```

```python
#Substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace = True)
```

```python
#Apagando as linhas com valores nulos
df.dropna(inplace = True)
```

```python
#Apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace = True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace = True)
```

**Criando colunas novas**

```python
#Criando a coluna de receita mul=multiplicação
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#nlargest = escolhe o top 3 com base na coluna receita, 3 linhas com maiores receitas
df.nlargest(3, "Receita")
```

```python
#nsamllest = escolhe os 3 piores com base na coluna receita, 3 linhas com menores receitas
df.nsmallest(3, "Receita")
```

```python
#Agrupando por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

**Trabalhando com datas**

```python
#Tranformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#Verificar o tipo de dado de cada coluna
df.dtypes
```

```python
#Transformando coluna de data em data = pd.to_datetime
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupando por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma coluna nova com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mês e o dia dt.month= mes / dt.day= dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data"].min()
```

```python
#Calculando a diferença de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#Criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
vendas_marco_19
```
